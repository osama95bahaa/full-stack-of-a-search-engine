```python
import os
import elasticsearch

es = elasticsearch.Elasticsearch() 
for root, dirs, files in os.walk(r'crawledData'):
    for file in files[:]:
        with open(os.path.join(root,file), 'r') as f:
            fnamesplit = file.split("-")
            line = f.readlines()
            es.index(index='scrapped', doc_type='blog', id= fnamesplit[0], body={
    'author': line[0],
    'url': line[1],
    'title': fnamesplit[1],
    'article': line[3:]
})
```

```python
#search by word in title.

titleInput = input('Please type a word to search in title.')
es.search(index="scrapped", body={"query": {"match": {"title" : titleInput}}})
```

```python
#Search by word in author name

authorInput = input('please type a word to search in authors.')
es.search(index="scrapped", body={"query": {"match": {"author" : authorInput}}})
```

```python
#Search by word in article

articleInput = input('please type a word to search in articles.')
es.search(index="scrapped", body={"query": {"match": {"article" : articleInput}}})
```

```python
#search for fuzzy query

authorFuzzyInput = input('please type a fuzzy word to search in author.')
es.search(index="scrapped", body={"query": {"fuzzy": {"author" : authorFuzzyInput}}})

# titleFuzzyInput = input('please type a fuzzy word to search in title.')
# es.search(index="scrapped", body={"query": {"fuzzy": {"title" : titleFuzzyInput}}})

# articleFuzzyInput = input('please type a fuzzy word to search in article.')
# es.search(index="scrapped", body={"query": {"fuzzy": {"article" : articleFuzzyInput}}})
```

```python
#search for multiple queries

authorTitleInput1 = input('1stQuery: please type a word to search in author.')
authorTitleInput2 = input('2ndQuery: please type a word to search in title.')
es.search(index="scrapped", body={"query": {"bool" : {"must": [ {"match" : {"author" : authorTitleInput1}},
                                                                {"match" : {"title" : authorTitleInput2}}]}}})

# authorTextInput1 = input('1stQuery: please type a word to search in author.')
# authorTextInput2 = input('2ndQuery: please type a word to search in text.')
# es.search(index="scrapped", body={"query": {"bool" : {"must": [ {"match" : {"author" : authorTextInput1}},
#                                                                 {"match" : {"article" : authorTextInput2}}]}}})

# titleTextInput1 = input('1stQuery: please type a word to search in title.')
# titleTextInput2 = input('2ndQuery: please type a word to search in text.')
# es.search(index="scrapped", body={"query": {"bool" : {"must": [ {"match" : {"title" : titleTextInput1}},
#                                                                 {"match" : {"article" : titleTextInput2}}]}}})
```
